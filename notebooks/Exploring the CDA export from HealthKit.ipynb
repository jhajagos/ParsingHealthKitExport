## Exploring the CDA exported from HealthKit

Apple allows you to export your HealthKit data from your iPhone. See for example: http://www.ryanpraski.com/apple-health-data-how-to-export-analyze-visualize-guide/. 

This workbook is to explore how to extract from the XML file the needed data

```python
import xml.etree.ElementTree as et
import os
```

```python
data_directory = "C:\\Users\\janos\\data\\apple_health_export\\"
```

```python
cda_filename = os.path.join(data_directory, "export_cda.xml")
```

```python
cda = et.parse(cda_filename)
```

```python
root = cda.getroot()
```
